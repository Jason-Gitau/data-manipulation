```python
# loading the data
import pandas as pd

df=pd.read_csv(r'C:\Users\Jason\Documents\personal work\3month course on ML\data manipulation\archive\Automobile_data.csv')
df
```

```python
df.info()
```

```python
# look for missing values
df.isnull().sum()
# ass we can see there are no missing values
```

```python
# Replace "?" with NaN and convert ['normalized-losses'] column to numerical values
df['normalized-losses'].replace("?", pd.NA, inplace=True)
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'])
df
```

```python
# handling duplicates
# step 1: check if there any duplicates
duplicates=df.duplicated().sum()
# apparently, there are no duplicated, but if there there were this is what we would have done.
if duplicates>0:
    df.drop_duplicates(inplace=True)
else:
    print('There are no duplicates')

```

```python
# handle outliers
# here we are replacing ? with NaN and converting other values into numeric for the quartile method to work
df['price']=df['price'].replace('?',pd.NA)
df['price']=pd.to_numeric(df['price'])

# this is where handling outliers actually starts
q1=df['price'].quantile(0.25)
q3=df['price'].quantile(0.75)
IQR=q3-q1
lower=q1-1.5*IQR
upper=q3+1.5*IQR

df=df[(df['price']>=lower)&(df['price']<=upper)]

```
